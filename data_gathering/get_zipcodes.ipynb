```python
from bs4 import BeautifulSoup
import urllib.request
import re
```

```python
nova_counties = ['Arlington', 'Fairfax', 'Loudoun', 'Prince William','Stafford','Alexandria City', 'Fairfax City', 'Falls Church City', 'Manassas City','Manassas Park City']
```

```python
URL = 'https://www.zipcodestogo.com/Virginia/'
page = urllib.request.urlopen(URL)
soup = BeautifulSoup(page,'html.parser')
```

```python
print(soup.prettify())
```

```python
zipcode_table = soup.find('table',attrs={'class':'inner_table'})
data = zipcode_table.find_all('tr')
```

```python
rows = [x.find_all('td') for x in data]
```

```python
for i,item in enumerate(rows[1]):
    print(i,item)
```

Starting with row 2, the zip code is in cell 0 and the county is in cell 2

```python
z =str(rows[2][0].find_all('a')[0])
zc = str(z).split('>')[1].split('<')[0]
print(zc)
```

```python
good_zips = []
for row in rows[2:]:
    # check if this row is from one of the NoVA counties
    ck = [county for county in nova_counties if(county in str(row[2]))]
    if bool(ck):
        z = str(row[0].find_all('a')[0])
        zc = str(z).split('>')[1].split('<')[0]
        good_zips.append(zc)
```

```python
print(good_zips)
```

```python
import csv
with open('zipcode_list.csv','w') as csv_out:
    zc_writer = csv.writer(csv_out)
    for code in good_zips:
        zc_writer.writerow([code])
```

```python
test = []
with open('zipcode_list.csv','r') as csv_in:
    zc_reader = csv.reader(csv_in)
    for row in zc_reader:
        test.append(row)
```
